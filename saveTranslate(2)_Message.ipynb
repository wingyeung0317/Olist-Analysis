```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("E:/JDE/olist/olist_order_reviews_dataset.csv")
```

```python
from textblob import TextBlob
```

```python
for i, comment in enumerate(dataset['review_comment_message']):
    try:
        translated_comment = TextBlob(comment).translate(from_lang='pt-BR', to='en')
    except:
        translated_comment = comment
    dataset.loc[i, 'translate_comment_message'] = str(translated_comment)
    # if i == 7:
    #     break
```

```python
dataset
```

```python
DF_save = dataset[['review_id', 'order_id', 'translate_comment_message']]
DF_save.to_csv('E:/JDE/olist/comment_message_translate.csv')
```

```python
def sentiment_analysing(text):
    try:
        score = TextBlob(text).sentiment.polarity
    except:
        score = np.nan
    return score

DF_save['message_sentiment'] = DF_save['translate_comment_message'].apply(lambda x: sentiment_analysing(x))
```
